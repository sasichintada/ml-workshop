####Basic Line Plot

Plot a simple line graph of ùë¶ = ùë•
2
for ùë• values ranging from -10 to 10.
. Label the x-axis as "X Values" and the y-axis as "Squared Values."

```python
import matplotlib.pyplot as plt
import numpy as np
```

####Bar Plot

####Histogram

Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
Customize the number of bins and add a grid to the plot

```python
#stimulating 100 random numbers
data=np.random.randn(100)
#creating a histogram with bins
plt.hist(data, bins=10, color='skyblue',edgecolor='black')
#labels and title for the plot
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Histogram')
#grid is used to simplify the plot
plt.grid(True)
#display the plot
plt.show()
```

####Pie Chart


Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
Display the percentage values on the chart and use a color palette for
differentiation.

```python
#data set with categories and values
categories=['A','B','C','D','E']
values=[20,30,25,15,10]
#colors for each part of pie
colors=['lightblue','pink','orange','red','maroon']
#plotting a pie chart
plt.pie(values, labels=categories, colors=colors)
#title for the pie chart
plt.title('Distribution of 5 categories')
#display
plt.show()
```

Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
 Highlight the differences in the code, syntax, and aesthetics of the two
libraries

```python
#dataset with quantity and price
data = {
    'Quantity': [10, 20, 30, 40, 50],
    'Price': [5, 6, 7, 8, 9]
}
df = pd.DataFrame(data)
```

####Matplotlib

```python
import matplotlib.pyplot as plt
# Line Plot
plt.plot(df['Quantity'], df['Price'],marker='o',color='green')
plt.title('matplotlib')
plt.xlabel('Quantity')
plt.ylabel('Price')
#display
plt.show()
```

#### Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Line Plot
sns.lineplot(x='Quantity', y='Price', data=df, marker='o', color='green')
plt.title('Seaborn')
plt.xlabel('Quantity')
plt.ylabel('Price')
#display
plt.show()
```

 **Differences in the code, syntax, and aesthetics of the two libraries**

**Code**

*   **Matplotlib :** It contains more lines of code and we have to set it manually
*   **Seaborn :**  Shorter code and most of the things have automatic functionality

**Syntax**


* **Matplotlib :** You need to use separate commands for labels and colors
* **Seaborn :**  Easier syntax with functions

**Aesthetics:**


*  **Matplotlib:** The plots are simple and plain.we need to add styling for it
*  **Seaborn:** By default the plots are with good colors,grids and formatting

###Save and Style Plots


 Plot any graph of your choice and save it in both .png and .pdf formats.
 Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization

```python
x = np.linspace(-10, 10, 400) #values of x
y = x**2   #value of y
plt.style.use('ggplot') #using ggplot style
plt.plot(x, y,label='x^2', linewidth=2)
#labels for x,y axis
plt.xlabel('X Values')
plt.ylabel('Y Values')
#title for the plot
plt.title('y = x^2')
plt.legend()
# plots as png and pdf
plt.savefig('x_squared_plot.png')
plt.savefig('x_squared_plot.pdf')
#display the plot
plt.show()
```
